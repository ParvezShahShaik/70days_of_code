# Matplotlib WorkFlow

```python
#import the packages and [%matplotlib inline] ensure that the plots generated by matplotlib appear directly in the notebook.
%matplotlib inline 
import matplotlib.pyplot as plt

#1.preparing the data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

#2.setup the plot
fig, ax = plt.subplots(figsize= (5,5)) #width, height

#plot data
ax.plot(x, y) 

#4.customize plot
ax.set(title="simple plot",
      xlabel="x-axis",
      ylabel="y-axis");

#5. save and show(you save the whole figure)
#fig.savefig("images/sampleplot.png")
```

# Making figures with NumPY arrays

we create:
* Line plot
* Scatter Plot
* Bar plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
#create some data
x= np.linspace(0, 10, 100)
x[:10]
```

```python
#Plot the data and create a line plot
fig,ax = plt.subplots()
ax.plot(x, x**2);
```

```python
#create a ascatter plot with same data
fig,ax = plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
#Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
#Make a plot from dictionary 

nut_butter_prices = {"Almond butter": 10,
                    "Peanut butter": 9,
                    "Cashew butter": 12}
fig,ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Dan's nut butter store",
      ylabel="Price ()");
```

```python
#Horizontal bars

fig,ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make some data and create Histograms

x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

# Two options for Subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows= 2,
                                            ncols= 2,
                                            figsize= (10,5))
# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
#ax3.set(title="Dan's nut butter store",
      #ylabel="Price ()");
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2

fig, ax = plt.subplots(nrows = 2,
                      ncols = 2,
                      figsize = (10,5));

#Plot to each different axis
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```
